```python
def update_table(
    dataframe,
    table,
    database="cup_adventure",
    username=os.getenv("AWS_USER"),
    passwd=os.getenv("AWS_PASSWORD"),
    hostname=os.getenv("AWS_HOST"),
    portnum=int(os.getenv("AWS_PORT")),
):
    # Create Connection
    engine = create_engine(f"mysql://{username}:{passwd}@{hostname}:{portnum}/{database}")
    connection = engine.connect()
    # Update table
    try:
        dataframe.to_sql(
            table, connection, if_exists="replace", index=False
        )
        print(f"Success: {table} updated in database")
    except:
        print(f"Error: {table} could not be updated in database")
    pass
```
